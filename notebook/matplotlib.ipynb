```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\khati\\GITHUB\\python\\dataset\\netflix_titles_2021.csv")
```

#### Top 10 Countries Producing Netflix Content

```python
# Get top 10 countries with the most content
top_countries = df["country"].value_counts().head(10)

plt.bar(top_countries.index, top_countries.values, color='skyblue')
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.title("Top 10 Countries Producing Netflix Content")
plt.xticks(rotation=45)
plt.show()
```

#### Distribution of Movies vs. TV Shows

```python
# Count the occurrences of each type
type_counts = df["type"].value_counts()

plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=["red", "blue"], startangle=90)
plt.title("Distribution of Movies vs. TV Shows on Netflix")
plt.show()
```

####  Number of Releases Per Year

```python
plt.hist(df["release_year"], bins=100, color="green", edgecolor="black")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Number of Netflix Releases Per Year")
plt.show()
```

#### Trend of Netflix Releases Over the Years

```python
# Count titles released per year
year_counts = df["release_year"].value_counts().sort_index()
```

```python
year_counts
```

```python
# Plot
plt.figure(figsize=(12, 6))
plt.plot(year_counts.index, year_counts.values, marker="o", linestyle="-", color="purple")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.title("Trend of Netflix Releases Over the Years")
plt.grid(True)
plt.show()
```

#### Top 10 Directors with the Most Netflix Content

```python
top_directors = df["director"].value_counts().head(10)

plt.figure(figsize=(12, 6))
plt.barh(top_directors.index, top_directors.values, color='orange')
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.title("Top 10 Directors with the Most Netflix Content")
plt.show()
```

#### Relationship Between Release Year and Duration

```python
# Filter only movies
movies_df = df[df["type"] == "Movie"]

plt.figure(figsize=(10, 6))
plt.scatter(movies_df["release_year"], movies_df["duration"], alpha=0.5, color="red")
plt.xlabel("Release Year")
plt.ylabel("Duration (minutes)")
plt.title("Movie Duration Over Time")
plt.show()
```
